```python
from time import time
import random as rand
import matplotlib.pyplot as plt
from collections import deque as linked_list
```

```python
def InsertionSort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1

        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

def PrintArray(arr):
    for x in range (0, len(arr)):
        print(arr[x])
```

```python
def BucketSort(arr):
    bucket = [x for x in range (0, 10)]
    for i in range (0, len(bucket)):
        bucket[i] = linked_list()
    for i in range (0, len(arr)):
        bucket[int(arr[i]/10)].append(arr[i])
    arr = []
    for i in range (0, len(bucket)):
        InsertionSort(bucket[i])
        arr += bucket[i]
    return arr
```

```python
def TimeBucketSort(arr):
    start_time = time()
    BucketSort(arr)
    return  time() - start_time
```

```python
#Testing if the bucketsort works correctly
arr = [23, 1, 10, 5, 2, 15, 25]
print(TimeBucketSort(arr))
arr = BucketSort(arr)
PrintArray(arr)
```

```python
n = [x for x in range (0, 1001)] #Creating arrays of different sizes from 0 to 1000
evenly_spaced_times = [x for x in range (0, 1001)] #Fill the arrays evenly
unevenly_spaced_times = [x for x in range (0, 1001)] #Fill the arrays unevenly

#Testing the time of sorting for evenly
for i in range (0, len(n)):
    array = [x for x in range (0, i)]
    for j in range (0, len(array)):
        array[j] = rand.randint(0, 99)
    evenly_spaced_times[i] = TimeBucketSort(array)

#Testing the time of sorting for unevenly
for i in range (0, len(n)):
    array = [x for x in range (0, i)]
    for j in range (0, len(array)):
        array[j] = rand.randint(0, 99)
    unevenly_spaced_times[i] = TimeBucketSort(array)

#Showing the graphplt.figure()
plt.plot(n, evenly_spaced_times, '.b')
plt.plot(n, unevenly_spaced_times, 'o', color='orange')
plt.show()
```
